### Gold Layer View : Overall Sentiment Share

```python
%sql
CREATE OR REPLACE VIEW sentimental_analysis.gold.overall_sentiment_share AS
SELECT
  sentiment_label,
  COUNT(*) AS total_count,
  ROUND(
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (),
    2
  ) AS percentage
FROM sentimental_analysis.gold.gold_ml_core
GROUP BY sentiment_label;
```

### Gold layer View : Overall Emotion Share

```python
%sql
CREATE OR REPLACE VIEW sentimental_analysis.gold.overall_emotion_share AS
SELECT
  emotion_label,
  COUNT(*) AS total_count,
  ROUND(
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (),
    2
  ) AS percentage
FROM sentimental_analysis.gold.gold_ml_core
GROUP BY emotion_label;
```

### Gold Layer View : Daily Net Sentiment Score

```python
%sql
CREATE OR REPLACE VIEW sentimental_analysis.gold.daily_net_sentiment AS
SELECT
  created_date,
  SUM(CASE WHEN sentiment_label = 'positive' THEN 1 ELSE 0 END) -
  SUM(CASE WHEN sentiment_label = 'negative' THEN 1 ELSE 0 END)
  AS net_sentiment_score
FROM sentimental_analysis.gold.gold_ml_core
GROUP BY created_date;
```
